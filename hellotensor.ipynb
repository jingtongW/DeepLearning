```python
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score

from sklearn.model_selection import cross_val_score
```

```python
train = pd.read_csv('iris.csv')
```

```python
X = train.iloc[:,:4]
y = train.iloc[:,4:]
```

```python
print(len(X),len(y))
```

```python
y_keys = y['Iris-setosa'].unique()
```

```python
for i in range(len(y_keys)):
    y['Iris-setosa'] = y['Iris-setosa'].replace(y_keys[i], i)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
def ModelTraining(model):
    model.fit(X_train, y_train)
    predict_target = model.predict(X_test)
    accuracy = round(cross_val_score(model, X_train, y_train).mean()*100, 4)
    test_accuracy = accuracy_score(y_test, predict_target)
    print("Train_Accuracy : ", accuracy , "%")
    print("Test_Accuracy :", test_accuracy, "%")
    return predict_target
```
